**QUESTION 6:**    Explain the concept of generators in Python and how they are defined.

**ANSWER 6:** .Generators in Python are a special type of function used to iterate over data in a memory-efficient way. Unlike normal functions, generators do not return all values at once. Instead, they produce values one by one when required.

A generator pauses its execution after yielding a value and resumes from the same point when the next value is requested. This makes generators very efficient when working with large datasets.
in generator functions we use yield instead of return.

```python
#question 1:   Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
#answer#

def sum_even_numbers(a):
  total=0
  for i in a :
    if i%2==0:
      total=total+i
  return(total)

b=[2,3,4,5,6,7,8,9,10]
sum_even_numbers(b)

```

```python
#question 3: Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
#answer:

def square_of_list(s):
  n=[]
  for i in s:
    n.append(i**2)
  return(n)

lis=[2,3,4,5,6]
square_of_list(lis)
```

```python
#question 4: Write a Python function that checks if a given number is prime or not from 1 to 200.
#answer:

def is_prime(num):
  if num <= 1:
    return False

  for i in range(2, num):
    if num % i == 0:
      return False

  return True

is_prime(7)
```

```python
#question 5 : Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
#answer

def fibonacci_generator(num_terms):
  a=0
  b=1
  count = 0
  while count < num_terms:
    yield a
    a, b = b, a + b
    count += 1

print("First 5 Fibonacci numbers:")
for num in fibonacci_generator(5):
  print(num)


my_fib_gen = fibonacci_generator(5)
print(next(my_fib_gen))
print(next(my_fib_gen))
print(next(my_fib_gen))
print(next(my_fib_gen))
print(next(my_fib_gen))
```

```python
#question 6: Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power(n):
  for i in range(n+1):
    yield(2**i)

m=power(5)
print(next(m))
print(next(m))
print(next(m))
print(next(m))
print(next(m))
```

```python
#question 7:  Implement a generator function that reads a file line by line and yields each line as a string.
#answer:

def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line
```

```python
#question 8: Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

#answer:
data = [(1, 3), (4, 1), (2, 2), (5, 0)]

sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
```

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map.

```python
#answer 11:

orders =  [[34587, 4, 40.95],[98762, 5, 56.80],[77226, 3, 32.95],
 [88112, 3, 24.99]]

m=list(map(lambda x: (x[0], x[1]*x[2] + 10 if x[1]*x[2] < 100 else x[1]*x[2]),
        orders))


print(m)
```
